# Optimizing params for the residual function -- Whittle likelihood

Fitting in frequency domain using Whittle likelihood

$$
ℓ_W(\theta) = -\sum\limits _{\omega\in\Omega}\begin{Bmatrix}\log\hat{f}(\omega;\theta)+\frac{I(\omega)}{\hat{f}(\omega;\theta)}\end{Bmatrix}  
$$

where $\hat{f}(\omega;\theta)$ is the continuous-time process spectral density (what we expected)  and $I(\omega)$ is the periodogram (what we observed)

## Select Frequency range to omit the influence of inertial and M4 peak 

Generally, the Whittle likelihood is applied to optimize the parameters for the whole frequency band of the observed periodogram (an estimation of the true spectrum). However, in our case, we selected a range of frequencies (from coriolis frequency to buoyancy frequency) to perform the optimization because the internal tide behaviors appear in this bandwidth. In addition, since the inertial and M4 peak are not the characteristics of the internal tides, they are omitted in our optimization using Whittle likelihood.

```python
# select the frequency range from corilolis freq to buoyancy freq, and omit M4 frequency
def subset_IT(ff_list,coriolis_freq):
    f_coriolis = coriolis_freq
    coriolis_peak_bandwidth = 0.1
    M4_peak_bandwidth  = 0.1
    M4_freq = 2*0.0805*24 #cpd
    buoyancy_freq = 35
    
    subset_list = []
    for i in range(len(ff_list)):
        ff = ff_list[i]
        subset = []
        for ii in range(len(ff)):
            subset_range = np.logical_or(np.less(f_coriolis + coriolis_peak_bandwidth, ff[ii]) & np.less(ff[ii], M4_freq-M4_peak_bandwidth), #omit M4 peak
                                         np.less(M4_freq + M4_peak_bandwidth, ff[ii]) & np.less(ff[ii],buoyancy_freq)) 
            subset.append(subset_range)
        subset_list.append(subset)
    return subset_list
```

```python
subset_p_P1 = subset_IT(F_p_P1,f_coriolis_cpd)
subset_w_p1 = subset_IT(F_w_P1,f_coriolis_cpd)

F_p_P1_modulated = processing.select_frequency(F_p_P1,subset_p_P1 )
P_p_P1_modulated = processing.select_frequency(P_p_P1,subset_p_P1 )
```

## M1L2 Kernal

```python
def M1L2(dx,params):
    
    noise_var = params[0]
    η_matern1 = params[1]
    α_matern1 = params[2]
    eta1      = params[3]
    ℓ_exp1    = params[4]
    eta2      = params[5]
    ℓ_exp2    = params[6]
    
    matern1 = cov.matern(dx, (η_matern1,α_matern1),lmbda=20,sigma=1e-6)              #background energy continuum  
    peak1 = cov.LR(dx,(eta1,ℓ_exp1),l_cos=1/((O1_freq+K1_freq)/2))
    peak2 = cov.LR(dx,(eta2,ℓ_exp2),l_cos=1/((S2_freq+N2_freq)/2))
    noise = cov.white_noise(dx,noise_var)
    COV = matern1 + peak1 + peak2 + noise
    return COV
```

```python
def M1L2_no_noise(dx,params):
    
#     noise_var = params[0]
    η_matern1 = params[0]
    α_matern1 = params[1]
    eta1      = params[2]
    ℓ_exp1    = params[3]
    eta2      = params[4]
    ℓ_exp2    = params[5]
    
    matern1 = cov.matern(dx, (η_matern1,α_matern1),lmbda=20,sigma=1e-6)              #background energy continuum  
    peak1 = cov.LR(dx,(eta1,ℓ_exp1),l_cos=1/((O1_freq+K1_freq)/2))
    peak2 = cov.LR(dx,(eta2,ℓ_exp2),l_cos=1/((S2_freq+N2_freq)/2))
#     noise = cov.white_noise(dx,noise_var)
    COV = matern1 + peak1 + peak2 #+ noise
    return COV
```

```python
def subset_diurnal_peak(ff_list):
    peak_bandwidth = 0.5
    M1_freq = 1 #cpd
    
    subset_list = []
    for i in range(len(ff_list)):
        ff = ff_list[i]
        subset = []
        for ii in range(len(ff)):
            subset_range = np.less(M1_freq - peak_bandwidth,ff[ii]) & np.less(ff[ii],M1_freq + peak_bandwidth)
            subset.append(subset_range)
        subset_list.append(subset)
    return subset_list
```

```python
def subset_semidiurnal_peak(ff_list):
    peak_bandwidth = 0.5
    M2_freq = 2 #cpd
    
    subset_list = []
    for i in range(len(ff_list)):
        ff = ff_list[i]
        subset = []
        for ii in range(len(ff)):
            subset_range = np.less(M2_freq - peak_bandwidth,ff[ii]) & np.less(ff[ii],M2_freq + peak_bandwidth)
            subset.append(subset_range)
        subset_list.append(subset)
    return subset_list
```

```python
def cal_whittle_value(P_obs_list, Time_obs_list, subset_p_list,soln_model_fit_list, kernel):
    whittle_value_list = []
    for i in range(len(P_obs_list)):
        Time = Time_obs_list[i]
        Subset = subset_p_list[i]
        P_obs = P_obs_list[i]
        params = soln_model_fit_list[i]
        whittle_value = []
        
        for window in range(len(P_obs)):
            n = len(Time[window])
            delta = (Time[0][1]-Time[0][0]).astype('float')/1e9/86400
            data_length = len(P_obs[window][Subset[window]])
            value = optimisation.whittle_function(params[window],n,delta,Subset[window],P_obs[window],kernel)
            value_per_point = value/data_length  #per points
            whittle_value.append(value_per_point)
               
        whittle_value_list.append(whittle_value)
                                 
    return whittle_value_list
```

```python
M1L1_Solution_list = change_window_label([
    Soln_P10_M1L1,
    Soln_P11_M1L1,
    Soln_P12_M1L1,
    Soln_P13_M1L1,
    Soln_P14_M1L1,
    Soln_P15_M1L1,
    Soln_P16_M1L1,
    Soln_P17_M1L1,], Start_date_label_list, End_date_label_list)

Season_label_list =  ['Feb-Apr','May - Jul','Aug - Oct','Nov-Jan',
                      'Feb-Apr','May - Jul','Aug - Oct','Nov-Jan',
                      'Feb-Apr','May - Jul','Aug - Oct','Nov-Jan',
                      'Feb-Apr','May - Jul','Aug - Oct','Nov-Jan',]

solution_df_list = []
for order, solution in enumerate(M1L1_Solution_list):
    solution_df = create_df(solution,parameter_name=M1L1_parameter_name)
    solution_df.set_index('start_date',inplace=True)
    # in this case, we have 5 solutions and each each solution contains 2 windows
    selected_rows1 = solution_df.loc[Start_date_label_list[2*order]]
    selected_rows2 = solution_df.loc[Start_date_label_list[2*order+1]]
    selected_rows1['end_date'] = End_date_label_list[2*order]
    selected_rows1['mean_temp'] = Temp_avg_df_list[2*order]['mean_temp'] 
    selected_rows1['season']   = Season_label_list[2*order]
    selected_rows2['end_date'] = End_date_label_list[2*order+1]
    selected_rows2['mean_temp'] = Temp_avg_df_list[2*order+1]['mean_temp'] 
    selected_rows2['season']   = Season_label_list[2*order+1]
    
    selected_rows1.reset_index(inplace=True)
    selected_rows2.reset_index(inplace=True)
    
    solution_df_list.append(selected_rows1)
    solution_df_list.append(selected_rows2)
    
Soln_df_table_M1L1 = pd.concat(solution_df_list, axis=0) 
Soln_df_table_M1L1['site'] = 'ITFTIS'
Soln_df_table_M1L1['eta1'] = np.nan
Soln_df_table_M1L1['ℓ_exp1'] = np.nan
Soln_df_table_M1L1['Model_type'] = 'M1L1'
Soln_df_table_M1L1['depth'] = -Soln_df_table_M1L1['depth']
Soln_df_table_M1L1['Whittle_value'] = -Soln_df_table_M1L1['Whittle_value']
Soln_df_table_M1L1['start_date'] = pd.to_datetime(Soln_df_table_M1L1['start_date'])
Soln_df_table_M1L1['end_date'] = pd.to_datetime(Soln_df_table_M1L1['end_date'])
Soln_df_table_M1L1['year'] =  Soln_df_table_M1L1['start_date'].dt.year
desired_order = ['site','depth','year','start_date','end_date','season','mean_temp','Model_type','η_matern1','α_matern1','eta1','ℓ_exp1','eta2','ℓ_exp2','Whittle_value']
Soln_df_table_M1L1 = Soln_df_table_M1L1[desired_order]
Soln_df_table_M1L1.reset_index(drop=True,inplace=True)

Soln_df_table_M1L1.to_csv('ITFTIS M1L1 Model fit results.csv', index=False)
Soln_df_table_M1L1
```
