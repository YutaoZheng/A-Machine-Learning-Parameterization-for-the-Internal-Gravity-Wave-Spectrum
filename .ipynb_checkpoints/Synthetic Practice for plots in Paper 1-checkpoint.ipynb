```python
# Calcualte the analytical result for the brodening peak spectrum
def LR_sykulski_spectra(f, params, ℓ_cos1=0.5):
    eta, ℓ_exp1 = params
    ω_0 = 2 * np.pi / ℓ_cos1
    ω = 2 * np.pi * f - ω_0
    c = 1 / l_exp1
    return np.power(eta, 2.) / (np.power(ω, 2.) + np.power(c, 2.))


def LR_spectra(f, params, ℓ_cos1=0.5):
    eta, ℓ_exp1 = params
    b = 2 * np.pi / ℓ_cos1
    c = ℓ_exp1
    return np.power(eta, 2.) / c * (
        1 / (np.power(2 * np.pi * f - b, 2) + 1 / np.power(c, 2)) + 1 /
        (np.power(2 * np.pi * f + b, 2) + 1 / np.power(c, 2)))


def matern_spectra(ff, params, lmbda=20):
    """General Matern PSD a la Lilly/Sykulski"""

    eta = params[0]
    alpha = params[1]
    #     lmbda = params[2]

    c = (gamma(1 / 2) * gamma(alpha - 1 / 2)) / (2 * gamma(alpha) * np.pi)
    S = np.power(eta, 2) * np.power(lmbda, 2 * alpha - 1) / c
    S *= np.power(np.power(2 * np.pi * ff, 2) + np.power(lmbda, 2), -alpha)

    return S


def M1L1_spectra(f, params):
    noise = params[0]
    params_matern1 = params[1:3]
    params_LR1 = params[3:5]
    return matern_spectra(f, params_matern1, lmbda=20) + LR_spectra(
        f, params_LR1, ℓ_cos1=0.5) + np.power(
            np.abs(noise), 2)  #starts decay at semidiurnal peak


def M1L1_sykulski_spectra(f, params):
    noise = params[0]
    params_matern1 = params[1:3]
    params_LR1 = params[3:5]
    return matern_spectra(f, params_matern1, lmbda=20) + LR_sykulski_spectra(
        f, params_LR1, ℓ_cos1=0.5) + np.power(
            np.abs(noise), 2)  #starts decay at semidiurnal peak


def M1L2_spectra(f, params):
    noise = params[0]
    params_matern1 = params[1:3]
    params_LR1 = params[3:5]
    params_LR2 = params[5:7]
    return matern_spectra(f, params_matern1, lmbda=20) + LR_spectra(
        f, params_LR1,
        ℓ_cos1=1) + LR_spectra(f, params_LR2, ℓ_cos1=0.5) + np.power(
            np.abs(noise), 2)  #starts decay at semidiurnal peak
```

```python
#define acf
def LR_sykulski(dx,params,l_cos=2):
    
    eta = params[0]
    l_exp = params[1]
    
    ω_0 = l_cos/(2*np.pi)
    LR_cos = np.exp(np.array([complex(0, imag) for imag in dx])*ω_0)
    c = 1/l_exp
    LR_exp = np.exp(-np.abs(dx)/l_exp)
    return np.power(eta,2.)/(2*c)*LR_cos*LR_exp

#define acf
def LR(dx,params,l_cos=2):
    
    eta = params[0]
    l_exp = params[1]
    
    ω_0 = 2*np.pi*l_cos
    LR_cos = np.cos(dx*ω_0)
    
    LR_exp = np.exp(-np.abs(dx)/l_exp)
    return np.power(eta,2.)*LR_cos*LR_exp

# from speccy import acf
def matern(dx, params,lmbda=10, sigma = 0, acf = True):
    """General Matern covariance a la Lilly/Sykulski"""

    eta = params[0]
    alpha = params[1]
#     lmbda = params[2]

    nu = alpha - 1/2
    
    K = 2 * np.power(eta, 2) / (gamma(nu) * np.power(2, nu))
    K *= np.power(np.abs(lmbda * dx), nu)
    K *= K_nu(nu, np.abs(lmbda * dx))
    K[np.isnan(K)] = np.power(eta, 2.)

    if acf:
        K[0] = K[0] + sigma**2
    else:
        n = dx.shape[0]
        K += sigma**2 * np.eye(n)
    
    return K

def white_noise(dx,var):
    return var*scipy.signal.unit_impulse(len(dx)) #delta function
                                         
def M1L1(dx,params):
    
    noise_var = params[0]
    η_matern1 = params[1]
    α_matern1 = params[2]
    eta1      = params[3]
    ℓ_exp1    = params[4]
    
    matern1 = matern(dx, (η_matern1,α_matern1),lmbda=20,sigma=1e-6)              #background energy continuum  
    peak1 = LR(dx,(eta1,ℓ_exp1),l_cos=0.5)
    noise = white_noise(dx,noise_var)
    COV = matern1 + peak1 + noise
    return COV

def M1L1_sykulski(dx,params):
    
    noise_var = params[0]
    η_matern1 = params[1]
    α_matern1 = params[2]
    eta1      = params[3]
    ℓ_exp1    = params[4]
    
    matern1 = matern(dx, (η_matern1,α_matern1),lmbda=20,sigma=1e-6)              #background energy continuum  
    peak1 = LR_sykulski(dx,(eta1,ℓ_exp1),l_cos=0.5)
#     peak1 = LR(dx,(eta1,ℓ_exp1),l_cos=0.5)
    noise = white_noise(dx,noise_var)
    COV = matern1 + peak1 + noise
    return COV


def M1L2(dx,params):
    
#     noise_var = params[0]
    η_matern1 = params[0]
    α_matern1 = params[1]
    eta1      = params[2]
    ℓ_exp1    = params[3]
    eta2      = params[4]
    ℓ_exp2    = params[5]
    
    matern1 = matern(dx, (η_matern1,α_matern1),lmbda=10,sigma=1e-6)              #background energy continuum  
    peak1 = LR(dx,(eta1,ℓ_exp1),l_cos=1)
    peak2 = LR(dx,(eta2,ℓ_exp2),l_cos=0.5)
#     noise = white_noise(dx,noise_var)
    COV = matern1 + peak1 + peak2 
    return COV
```

```python
def Matern_PSD_1st_derivative_true(ff, params , lmbda=10):
    eta = params[0]
    alpha = params[1]
    ω = 2*np.pi*ff
    
    S = -alpha*(2*ω)/(np.power(ω,2)+np.power(lmbda,2))
    
    return S
```
